```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
x=np.linspace(-10,10,100)
```

```python
fig,axs=plt.subplots(1,1,figsize=(8,8))
axs.plot(x,softmax(x))
axs.set_title('softmax')
axs.set(xlabel='x',ylabel='y')
```
